```
import pandas as pd
from google.colab import files
```

```
movies = pd.read_csv('movies.csv')
movies.head(10)
```

```
new = movies["title"].str.extract(r'(\([0-9]{4}\))$') 
movies["year"]= new 
movies.head()
```

```
new2 = movies["title"].str.split(r'\([0-9]{4}\)$',expand=True) 
movies["title"]= new2[0]
movies.head()
```

```
movies['year'] = movies['year'].str.replace(r'(\(|\))','',regex=True)
movies.head()
```

```
movies.dtypes
```

```
movies.shape
```

```
movies = movies.join(movies['genres'].str.get_dummies('|'))
movies.head(10)
```

```
movies.head()
```

```
links = pd.read_csv('links.csv')
links.head(10)
```

```
links.dtypes
```

```
links.shape
```

```
df_merged = links.merge(movies, on="movieId", how = 'inner')
```

```
df_merged.shape
```

```
df_merged.head()
```

```
ratings = pd.read_csv('ratings.csv')
ratings.head(10)
```

```
ratings.dtypes
```

```
ratings.shape
```

```
ratings_grouped = ratings.groupby('movieId',as_index=False)['rating'].mean()
ratings_grouped.head(25)
```

```
ratings_grouped.shape
```

```
tags = pd.read_csv('tags.csv')
tags.head()
```

```
tags.dtypes
```

```
tags.shape
```

```
tags=tags.groupby('movieId').agg({'tag': lambda x: ', '.join(x)})
tags.shape
```

```
tags=tags.reset_index()
```

```
tags.head()
```

```
df_merged2 = df_merged.merge(ratings_grouped, on="movieId", how = 'inner')
df_merged2.head(15)
```

```
df_final = df_merged2.merge(tags, on="movieId", how = 'left')
```

```
df_final.head()
df_final.shape
```

```
df_final["tag"].fillna("No tags available", inplace = True)  
df_final.head()
```

```
df_final.to_csv('movies_final.csv',index= False)
files.download('movies_final.csv')
```
